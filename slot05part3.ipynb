```python
import pandas as pd

year = [2010,2011,2012,2013,2014,2015]

salaryHCM = [8.5,9.0,10.5,12.0,11.0,10.0]

salaryLA = [5.5,6.0,6.0,5.5,6.0,5.0]

df = pd.DataFrame(list(zip(year, salaryHCM,salaryLA)),
                  columns=['Year', 'HCM','LA'])
df
#Lương tb TPHCM và Long An
#Số đơn vị hàng triệu quá lơn, cần đổi qua miền giá trị nhỏ 
#hơn để tránh tràng số
```

```python
#Nomarlization
#PPhap Max Absolute
#Giá trị x mới bằng giá trị x cũ chia cho max(|x cũ|)(Scale dữ liệu)
#Sau khi scale, giả trị dữ liệu được bảo toàn, từ đó đưa giá trị vào mô hình học máy
max_la = df['LA'].abs().max()
df['macLA'] = df['LA'] / max_la

df
```

```python
max_la = df['HCM'].abs().max()
df['macHCM'] = df['HCM'] / max_la

df
```

```python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)
```

```python
#sau khi huẩn luyện thì bỏ lương vào để nó transfrom
vSc = scaler.transform([[4.5,5.75]])
vSc
```

```python
vR = scaler.inverse_transform([[0.5,0.8]])
vR
#Revears lại giá trị trc khi transform
```

```python
#2. Min Max Scaler
#scale Min max về đoạn [0,1]
min_la = df['LA'].min()
max_la = df['LA'].max()
df['mmcLA'] = (df['LA'] - min_la) / (max_la - min_la)

df
```

```python
min_la = df['HCM'].min()
max_la = df['HCM'].max()
df['mmcHCM'] = (df['HCM'] - min_la) / (max_la - min_la)

df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)
```

```python
#Giả sử HCM và LA lần lượt là 10 và 4.5
#thì scale qua bằng bao nhiêu
vSc = scaler.transform([[10.0,4.5]])
vSc
```

```python
#Standaviation
#Z-score
#nếu 1 người thu nhập 15tr ở HCM và 12tr ở long an thì ai sướng hơn
df[['HCM','LA']].describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])

scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)

```

```python
#Một người thu nhập 15tr ở HCM và 1 người 12tr ở HCM, ai sướng hơn

sV = scaler.transform([[15,12]])
sV
#z-score HCM = 4.1
#z-score LA = 16
#Vậy LA cách trung bình 16 lần, HCM cách TB 4.1 lần
#LA sống sướng hơn
```
