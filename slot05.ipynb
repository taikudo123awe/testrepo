```python
#Data Transformation
import pandas as pd

df = pd.DataFrame({'Math': [4, 4, 7, 8, 12, 13, 15, 18, 22, 23, 23, 25],
                   'Eng': [2, 5, 4, 7, 7, 8, 5, 4, 5, 11, 13, 8]})
df.shape
```

```python
#Chia làm 5 khoảng có số lượng phần tưởng trong từng khoảng xấp xỉ bằng nhau
df['bin1'] = pd.qcut(df['Math'], q=5)
df
```

```python
df['bin1'].value_counts()
```

```python
#chia khoảng theo giá trị nhập vào( 0 -> 2 là 20%)
df['bin2'] = pd.qcut(df['Math'], q=[0, .2, .4, .6, .8, 1])
df
```

```python
# chia khoảng và đánh nhãn theo phân vị
df['bin3'] = pd.qcut(df['Math'],
                           q=[0, .2, .5, .6, .8, 1],
                           labels=['A', 'B', 'C', 'D', 'E'])
df
```

```python
df['bin4'] = pd.cut(df['Eng'],
                      bins=[0, 5, 11, float('Inf')],
                      labels=['Bad', 'OK', 'Good'])
df
```

```python
df['bin4'].value_counts()
```
