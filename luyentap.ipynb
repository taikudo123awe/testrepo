1. Biến định lượng (numerical variables):
Các biến định lượng có thể được trực quan hóa bằng nhiều loại biểu đồ khác nhau, tùy thuộc vào mục đích phân tích.

Histogram (Biểu đồ tần suất):

Ý nghĩa: Được sử dụng để hiển thị phân phối của một biến định lượng. Nó cho thấy tần suất xuất hiện của các giá trị trong từng khoảng giá trị (bins).
Thư viện: matplotlib, seaborn (plt.hist(), sns.histplot()).
Box Plot (Biểu đồ hộp):

Ý nghĩa: Biểu đồ này cho phép hiểu được sự phân phối, trung vị, và phát hiện các giá trị ngoại lai (outliers) trong dữ liệu định lượng.
Thư viện: seaborn (sns.boxplot()).
Scatter Plot (Biểu đồ phân tán):

Ý nghĩa: Dùng để biểu diễn mối quan hệ giữa hai biến định lượng. Mỗi điểm trên biểu đồ biểu diễn một cặp giá trị.
Thư viện: matplotlib, seaborn (plt.scatter(), sns.scatterplot()).
Line Plot (Biểu đồ đường):

Ý nghĩa: Thể hiện sự thay đổi của một biến định lượng qua thời gian hoặc chuỗi liên tục các giá trị.
Thư viện: matplotlib, seaborn (plt.plot(), sns.lineplot()).
Violin Plot (Biểu đồ violin):

Ý nghĩa: Kết hợp giữa box plot và biểu đồ mật độ, cho thấy sự phân phối của một biến định lượng và sự khác biệt giữa các nhóm.
Thư viện: seaborn (sns.violinplot()).
2. Biến định tính (categorical variables):
Các biến định tính thường được trực quan hóa bằng các loại biểu đồ sau:

Bar Plot (Biểu đồ cột):

Ý nghĩa: Biểu diễn tần suất của các nhóm định tính. Chiều cao của các cột cho thấy số lượng hoặc tỷ lệ của từng nhóm.
Thư viện: matplotlib, seaborn (plt.bar(), sns.barplot()).
Count Plot (Biểu đồ tần suất theo nhóm):

Ý nghĩa: Là phiên bản đơn giản của bar plot, biểu diễn tần suất của các giá trị trong một biến định tính.
Thư viện: seaborn (sns.countplot()).
Pie Chart (Biểu đồ tròn):

Ý nghĩa: Thể hiện tỷ lệ phần trăm của các giá trị định tính trong tổng thể. Tuy nhiên, biểu đồ tròn ít được sử dụng trong phân tích dữ liệu chuyên sâu.
Thư viện: matplotlib (plt.pie()).
3. Trực quan hóa kết hợp biến định lượng và định tính:
Khi muốn trực quan hóa mối quan hệ giữa biến định lượng và biến định tính, các biểu đồ dưới đây thường được sử dụng:

Box Plot với phân nhóm (Group by Categorical Variables):

Ý nghĩa: Giúp so sánh sự phân phối của biến định lượng dựa trên các nhóm của biến định tính.
Thư viện: seaborn (sns.boxplot(x=categorical_variable, y=numerical_variable)).
Violin Plot với phân nhóm:

Ý nghĩa: Hiển thị phân phối của biến định lượng cho từng nhóm định tính, cho phép so sánh dễ dàng.
Thư viện: seaborn (sns.violinplot(x=categorical_variable, y=numerical_variable)).
Strip Plot (Biểu đồ dải):

Ý nghĩa: Mỗi điểm dữ liệu được vẽ như một điểm phân tán, thể hiện mối quan hệ giữa biến định lượng và biến định tính.
Thư viện: seaborn (sns.stripplot()).
Point Plot (Biểu đồ điểm):

Ý nghĩa: Thể hiện trung bình của biến định lượng cho từng nhóm của biến định tính, kèm theo thanh lỗi (error bars) để thể hiện độ biến thiên.
Thư viện: seaborn (sns.pointplot()).

```python
#ta nhận thấy biến dân tộc chỉ có 3 giấ trị, tiến hành điền dữ liệu bị thiếu
df['DT'].fillna('KINH',inplace=True)
#dữ liệu đã không còn thiếu 
```

```python
#thao tác dữ liệu: tính điểm trung bình của 3 môn
df['TBM']=(df['M1']+df['M2']+df['M3'])/3
```

```python
df
```

```python
# phân loại các giá trị trong cột TBM
df.loc[df['TBM']<5.0,'XL'] = 'Fail'
df.loc[(df['TBM']>=5.0) & (df['TBM']< 7.0), 'XL'] ='Fair'
df.loc[(df['TBM']>=7.0) & (df['TBM']< 9.0), 'XL'] ='Good'
df.loc[df['TBM']>=9.0, 'XL'] ='excelence'
df
```

```python
df.loc[(df['TBM']>=5.0)&(df['TBM']<7),'XL'] = 'GT'
```

```python
dict_map = {    'A1':'G1', 'C':'G3', 'D1': 'G3', 'A': 'G1', 'B': 'G2'
}
df['NHOM'] = df['KT'].map(dict_map)
df
```

```python
def fplus(x,y):
    if(x=='G1' or x =='G2') and (y>=4.0):
        return 1.0
    return 0.0

df['PLUS'] = list(map(fplus,df['NHOM'],df['TBM']))
df
```
